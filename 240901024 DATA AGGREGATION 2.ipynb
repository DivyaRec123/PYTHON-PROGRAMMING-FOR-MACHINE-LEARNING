```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'A': ['sai', 'poo', 'sub', 'yuvi', 'rithi', 'sangu'],
'B': ['one', 'two', 'three', 'one', 'three', 'two'],
'C': np.random.randint(5, size=6),
'D': np.random.randint(5, size=6),
'E': np.random.randint(5, size=6)})
print("\nOriginal DataFrame\n",df)
```

```python
#Group by (multi-column): with one column sum
df1 = df.groupby(['A', 'B'], as_index=False)['C'].sum()
print("\nOriginal DataFrame\n",df)
print("\nGroup by (multi-column): with one column sum\n",df1)
```

```python
#Group by (multi-column): with one column sum
df1 = df.groupby(['A', 'B'], as_index=False)['C'].sum()
print("\nOriginal DataFrame\n",df)
print("\nGroup by (multi-column): with one column sum\n",df1)
```

```python
#combined groupby and aggregate function
df3 = df.groupby(['A', 'B'], as_index=False)['C'].agg('sum')
print("\nOriginal DataFrame\n",df)
print("\nCombined Groupby and Aggregate function\n",df3)
```

```python
#combined groupby and aggregate function – separate column headings
df4 = (df.groupby(['A', 'B']).agg([('average', 'mean'), ('total', 'sum')]))
print("\nOriginal DataFrame\n",df)
print("\nCombined Groupby and Aggregate function-separate column headings\n",df4)
```

```python
df5 = df.groupby(['A', 'B'], as_index=False).sum()
df6 = (df.groupby(['A', 'B']).agg(['sum']))
print("\nOriginal DataFrame\n",df)
print("\nGroupby-sum function\n",df5)
print("\nGroupby and Aggregate - sum function\n",df6)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9], [np.nan, np.nan, np.nan]], columns=['A', 'B', 'C'])
print("Original DataFrame\n",df)
print("Aggregate function: aggregate()\n", df.aggregate(['sum']))
print("Aggregate function: alias agg()\n", df.agg(['sum']))
```

```python
#Aggregate function to produce sum,min and max
a=df.agg(['sum', 'min', 'max'])
print("\nOriginal DataFrame\n",df)
print("\nAggregate function to produce sum, min and max\n",a)
```

```python
#Different aggregation per column
b=df.agg({'A': ['sum', 'min'], 'B': ['min', 'max']})
print("\nOriginal DataFrame\n",df)
print("\nDifferent aggregations per column\n",b)
```

```python
#Aggregate different functions over the columns and rename the index of the resulting DataFrame
c=df.agg(x=('A', max), y=('B', 'min'), z=('C', np.mean))
print("\nOriginal DataFrame\n",df)
print("\nAggregate different functions over the columns and rename the index of the resulting DataFrame\n",c)
```

```python
#Aggregate over the rows
d=df.agg("sum", axis="columns")
print("\nOriginal DataFrame\n",df)
print("\nAggregate-Sum: row wise\n",d)
```

```python
#Aggregate over the columns
e=df.agg("sum", axis="rows")
print("\nOriginal DataFrame\n",df)
print("\nAggregate-Sum: Column wise\n", e)
```

```python
#Summary of aggregate function using describe
print("\nOriginal DataFrame\n",df)
print("\nSummary of aggregate functions using describe()\n",df.describe())
```
